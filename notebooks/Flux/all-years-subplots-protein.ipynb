# Plotting the fluxes of organic matter in the Eastern Tropical North Pacific (ETNP) from high-resolution sediment trapping experiditions across years (2012-2019) and between three stations:

## - P1: -106.07344 W, 20.2636 N; coastal
## - P2: -106.85836 W, 16.85542 N; offshore
## - P3: -109.9001 W, 21.99014 N; offshore, N. ODZ with no secondary Chl maximum



```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# read the joined ctd and flux data files made in individual cruise notebooks (i.e., 2017-flux-subplots.ipynb)

P1_2017 = pd.read_csv("/home/millieginty/Documents/git-repos/2017-etnp/data/flux/SKQ_2017_P1_cast_trap_join.csv", index_col = [0])
P2_2017 = pd.read_csv("/home/millieginty/Documents/git-repos/2017-etnp/data/flux/SKQ_2017_P2_cast_trap_join.csv", index_col = [0])

P1_2018 = pd.read_csv("/home/millieginty/Documents/git-repos/2017-etnp/data/flux/RR_2018_P1_cast_trap_join.csv", index_col = [0])
P2_2018 = pd.read_csv("/home/millieginty/Documents/git-repos/2017-etnp/data/flux/RR_2018_P2_cast_trap_join.csv", index_col = [0])

P3_2019 = pd.read_csv("/home/millieginty/Documents/git-repos/2017-etnp/data/flux/KM_2019_P3_cast_trap_join.csv", index_col = [0])
P2_2019 = pd.read_csv("/home/millieginty/Documents/git-repos/2017-etnp/data/flux/KM_2019_P2_cast_trap_join.csv", index_col = [0])
                      
P2_2019.head()             
```

```python
# 3 panel plot with P2 org C fluxes in depth space
# For thesis talk, no letters and shaded ODZ

fig2, (ax2, ax4) = plt.subplots(1,2,sharey=True, figsize=(8,9))

#fig2.suptitle('Offshore Station P2 Organic carbon flux', fontsize=16, y = 1.0)

# 2017 P2

P2_2017.plot(ax = ax2, kind='scatter', x='Protein flux (ug/m2/day)', \
             y='Pressure [dbar]', c='Blue', zorder = 3, s = 50)
ax2.set_facecolor('#e0ecf4')
ax2.set_ylabel('Depth (m)', fontsize = 14)
ax2.set_xlim([0, 4000])
ax2.set_ylim(ax2.get_ylim()[::-1]) #this reverses the yaxis (i.e. deep at the bottom)
ax2.set_xlabel('Protein flux ($\mu$g m$^{-2}$ day$^{-1}$)', fontsize = 16, labelpad = 12)
ax2.xaxis.set_label_position('top') # this moves the label to the top
ax2.xaxis.set_ticks_position('top') # this moves the ticks to the top
ax2.tick_params(axis='x', labelsize=16)
ax2.tick_params(axis='y', labelsize=16)
ax2.grid(True)

ax2.axhline(y=75, color = 'black', linewidth=1) # horizonal line that denotes the onset of anoxia.

# 20%-grey rectangle spanning the axes' width from the ODZ y=75 to y=400
ax2.axhspan(75, 800, color='grey', alpha = 0.1)

# 20%-green rectangle spanning the axes' width from the DCM y=95 to y=120
ax2.axhspan(85, 120, color='green', alpha = 0.2)

ax2.annotate('Jan. 2017', xy=(150, 100), 
            xytext=(0.5, 0.3), textcoords='axes fraction',
            color='black', fontsize=16,
            horizontalalignment='center', verticalalignment='top'
            )

ax2.annotate('P2', xy=(150, 100), 
            xytext=(0.5, 0.2), textcoords='axes fraction',
            color='black', fontsize=16,
            horizontalalignment='center', verticalalignment='top'
            )


# 2019 P2

P2_2019.plot(ax = ax4, kind='scatter', x='Protein flux (ug/m2/day)', \
             y='Pressure [dbar]', c='Blue', zorder = 2, s = 50)
ax4.set_facecolor('#fff7bc') 
ax4.set_xlim([0, 4000])
ax4.set_xlabel('Protein flux ($\mu$g m$^{-2}$ day$^{-1}$)', fontsize = 16, labelpad = 12)
ax4.xaxis.set_label_position('top') # this moves the label to the top
ax4.xaxis.set_ticks_position('top') # this moves the ticks to the top
ax4.tick_params(axis='x', labelsize=16)
ax4.grid(True)
ax4.axhline(y=115, color = 'black', linewidth=1) # horizonal line that denotes the onset of anoxia.

# 20%-grey rectangle spanning the axes' width from the ODZ y=75 to y=400
ax4.axhspan(115, 780, color='grey', alpha = 0.1)

# 20%-green rectangle spanning the axes' width from the 120-145 m
ax4.axhspan(120, 145, color='green', alpha = 0.2)


ax4.annotate('Oct. 2019', xy=(150, 100), 
            xytext=(0.5, 0.3), textcoords='axes fraction',
            color='black', fontsize=18,
            horizontalalignment='center', verticalalignment='top'
            )

ax4.annotate('P2', xy=(150, 100), 
            xytext=(0.5, 0.2), textcoords='axes fraction',
            color='black', fontsize=18,
            horizontalalignment='center', verticalalignment='top'
            )


fig2.savefig('/home/millieginty/Documents/git-repos/2017-etnp/data/thesis-talk/entp-protein-p2.png', transparent=False, dpi=300, bbox_inches = "tight")

#ax4.annotate('C', xy=(30, 100), 
            #xytext=(0.94, 0.99), textcoords='axes fraction',
            #color='black', fontsize=18,
            #horizontalalignment='center', verticalalignment='top');
```
