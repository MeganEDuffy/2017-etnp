Starting with:

    Comet search results (.csv) of PTM-optimized database searches against ETNP-identidied + fungal proteins

Goal:

    Files with stripped (no PTMs) peptide lists and
    Columns with #'s of each modification in every sequence

## Calculating the false discovery rate (% FDR)

### Filtering PSMs > a selected XCorr value and exporting peptides

```python
# Let's separate out the decoy hits from the good ones

f_comet278pmm = f_comet278[~f_comet278['protein'].str.contains("DECOY")]
f_comet278dec = f_comet278[f_comet278['protein'].str.contains("DECOY")]

# how many PSM that are only PMM (proteins in the database)?

print("# real Comet PSMs", len(f_comet278pmm))

# compared to how many PSMs containing decoys?

print("# decoy Comet PSMs", len(f_comet278dec))

# calculate the bulk FDR (all PSMs so let's not beat ourselves up)

r = len(f_comet278pmm)
d = len(f_comet278dec)

FDR = d/r*100

print("False discovery rate = ", FDR)
```

```python
# keep only peptides  >3 XCorr
# need to convert Xcorr column from strings to numeric so we can use loc
# note that pmm here just means 'non decoy', it's a vestige from the accession prefixes from the MED proteins
f_comet278['xcorr'] = pd.to_numeric(f_comet278['xcorr'])

f_comet278_3 = f_comet278.loc[f_comet278['xcorr'] >= 2.278]

# What's the FDR?

# Let's separate out the decoy hits from the good ones

f_comet278pmm3 = f_comet278_3[~f_comet278_3['protein'].str.contains("DECOY")]
f_comet278dec3 = f_comet278_3[f_comet278_3['protein'].str.contains("DECOY")]

# how many PSM that are only PMM (proteins in the database)?

print("# real Comet PSMs", len(f_comet278pmm3))

# compared to how many PSMs containing decoys?

print("# decoy Comet PSMs", len(f_comet278dec3))

# calculate the FDR 

r = len(f_comet278pmm3)
d = len(f_comet278dec3)

FDR = d/(d+r)*100

print("False discovery rate = ", FDR)

# write modified dataframe to new txt file

f_comet278pmm3.to_csv("/home/millieginty/Documents/git-repos/2017-etnp/data/pro2020/ETNP-SKQ17-fungi-searches/SKQ17-Comet-fungi-searches/ETNP-SKQ17-278-265m-trap_PTMopt_Comet15_ETNPfungidb_00FDR.csv")
```
